## Criando Novas Váriaveis
Crie Variáveis [Valor Bruto, Valor m2, Valor Bruto m2]

```python
import pandas as pd
df = pd.read_csv('dados/aluguel_residencial_t.csv', sep=';')
df.head()
```

```python
df['Valor Bruto'] = df['Valor'] + df['Condominio'] + df['IPTU']
```

```python
df['Valor m2'] = df['Valor']/df['Area']
df.head()
```

```python
df['Valor m2'] = df['Valor m2'].round(2)
df.head()
```

```python
df['Valor Bruto m2'] = (df['Valor Bruto']/df['Valor m2']).round(2)
```

Váriavel Tipo Agregado - simplifica em casa ou apartamento os vários tipos de imóveis

```python
df['Tipo'].drop_duplicates()
```

```python
casa = ['Casa','Casa de Condomínio', 'Casa de Vila']
```

```python
def labda(x):
    if x in casa:
        return('Casa')
    else:
        return('Apartamento')
    
```

```python

df['Tipo Agregado'] = df['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
df.head()
#df['Tipo Agregado'] = df['Tipo'].apply(labda)
```
