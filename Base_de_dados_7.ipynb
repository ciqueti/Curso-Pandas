## Criando Agrupamentos

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial_2.csv', sep=';')
dados.head()
```

```python
dados['Valor'].mean().round(2)
# Média do 'Valor', mas e se quisermos a média por tipos de Bairro?
```

https://pandas.pydata.org/docs/reference/frame.html#computations-descriptive-stats

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados.head()
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
# Dicionario com nome e índices de recorrência
```

```python
for bairro, dados in grupo_bairro:
    print(dados)
# Para cada bairro temos um DataFrame associado
```

```python
for k, v in grupo_bairro:
    print(f'O Bairro - {k} possui a média de R$ {v["Valor"].mean().round(2)}')
```

```python
# grupo_bairro possui um grupo de DataFrames associados ao bairro que criamos, podemos:
grupo_bairro[['Valor','Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns={'min':'Mínimo', 'max':'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio de Aluguel por Bairro', {'fontsize' : 22})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio de Aluguel por Bairro', {'fontsize' : 22})
```
