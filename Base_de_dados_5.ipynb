## Tratamento de Dados Faltantes
Retirar Nan de 'Valor' do nosso DataFrame - Método dropna()

```python
import pandas as pd
```

```python
df = pd.read_csv('dados/aluguel_residencial.csv',sep=';')
df.head()
```

```python
df.info()
```

```python
df[df['Valor'].isnull()]
#Usando método isnull() selecionei uma series booleana para meu DataFrame
```

```python
A = df.shape[0]
df.dropna(subset=['Valor'],inplace=True)
B = df.shape[0]
A - B
```

## Tratamento de Dados - Continuação
Vamos retirar - Apartamentos com Condíminio nullo.
Em seguida adicione -  0 aos iptu e condominios restantes.

```python
df.head()
```

```python
#Condominios nullos
df[df['Condominio'].isnull()].shape[0]
```

```python
selecao = (df['Tipo'] == 'Apartamento') & (df['Condominio'].isnull() == True)
df = df[~selecao]
#selecao é os dados que queremos retirar, sendo assim mantemos seu oposto (~selecao)
```

```python
df[df['Condominio'].isnull()].shape[0]
```

```python
df.fillna({'Condominio' : 0 , 'IPTU' : 0},inplace=True)
```
